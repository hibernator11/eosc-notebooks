```python
import csv
from SPARQLWrapper import SPARQLWrapper

sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
```

Then we define our CONSTRUCT query to extract the metadata

```python
sparql.setQuery("""
PREFIX wdt: <http://www.wikidata.org/prop/direct/>
PREFIX wd: <http://www.wikidata.org/entity/>

CONSTRUCT {
  ?author wdt:P135 wd:Q667661 . 
  ?author wdt:P31 ?type .
  ?author wdt:P18 ?image .
  ?author wdt:P27 ?citizenship .
  ?author wdt:P569 ?dateBirth .
  ?author wdt:P19 ?placeBirth .
}
WHERE {
  ?author wdt:P135 wd:Q667661 . 
  ?author wdt:P31 ?type .
  OPTIONAL {?author wdt:P18 ?image .}
  OPTIONAL {?author wdt:P27 ?citizenship .}
  OPTIONAL {?author wdt:P569 ?dateBirth .  }   
  OPTIONAL {?author wdt:P19 ?placeBirth .}
}
"""
)
```

Finally, we serialise the result

```python
results = sparql.queryAndConvert()
print(results.serialize())
```

```python
with open("output/literary-realism.ttl", "w") as text_file:
    text_file.write(results.serialize())
```
